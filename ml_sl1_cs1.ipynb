```python
sns.countplot(df['label'],label="Count")
plt.show()
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
prediction_var = ['meanfreq', 'sd', 'median', 'Q25', 'Q75', 'IQR', 'skew', 'kurt',
       'sp.ent', 'sfm', 'mode', 'centroid', 'meanfun', 'minfun', 'maxfun',
       'meandom', 'mindom', 'maxdom', 'dfrange', 'modindx']
```

```python
train, test = train_test_split(df, test_size = 0.2)
print(train.shape)
print(test.shape)
```

```python
train_X = train[prediction_var]# taking the training data input
train_y=train.label# This is output of our training data
# same we have to do for test
test_X= test[prediction_var] # taking test data inputs
test_y =test.label   #output value of test dat
```

```python
import sklearn.metrics as metrics
logistic = LogisticRegression()
logistic.fit(train_X,train_y)
temp=logistic.predict(test_X)
print(metrics.accuracy_score(temp,test_y))
```
