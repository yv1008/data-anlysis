```python
import pandas as pd
import numpy as np
df = pd.read_csv('run_or_walk.csv');
df
```

```python
df.drop(df.columns[[0,1,2]], inplace = True, axis = 1)
```

```python
df
```

```python
from sklearn.naive_bayes import MultinomialNB
import seaborn as sns # this is used for the plot the graph
from sklearn.model_selection import train_test_split # to split the data into two parts
from sklearn import metrics # for the check the error and accuracy of the model
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn import svm    		# To import the svm classifier

import random
```

```python
data = df
from sklearn.preprocessing import LabelEncoder
encoded_data = data.apply(LabelEncoder().fit_transform)
X = x =df.drop("activity",axis=1)
y = df['activity']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.2,random_state=85)
```

```python
from sklearn.naive_bayes import GaussianNB
algorithm = GaussianNB(priors=None, var_smoothing=1e-9)
```

```python
algorithm.fit(X_train, y_train)
```

```python
print(algorithm.classes_)
```

```python
print('The Gaussian Model Has Achieved %.2f Percent Accuracy'%(algorithm.score(X, y)))
```

```python
y_predict = algorithm.predict(X_test)
y_predict
```

```python
accuracy = accuracy_score(y_test,y_predict)*100
accuracy
```
